```python
import pandas as pd
import numpy as np

df = pd.read_json("renthop_train.json")
df
```

```python
df['bathrooms_per_bedroom'] = df['bathrooms'] / df['bedrooms'].replace(0, np.nan)
df['bedrooms_per_bathroom'] = df['bedrooms'] / df['bathrooms'].replace(0, np.nan)
```

```python
df['price_per_bedroom'] = df['price'] / df['bedrooms']
df
```

```python
df['is_luxury'] = (df['bathrooms'] > 2) & (df['bedrooms'] > 3)
df['is_luxury'] = df['is_luxury'].map({True: "Yes", False: "No"})
```

```python
interest_price_map = df.groupby("interest_level")['price'].mean()
df['interest_price_mean'] = df['interest_level'].map(interest_price_map)
```

```python
df['total_rooms'] = df['bathrooms'] + df['bedrooms']
```

```python

latitude = df['latitude'].mean()
longtitude = df['longitude'].mean()
center = [latitude, longtitude]
```

```python
from geopy.distance import geodesic

df['dist_from_center'] = df.apply(
    lambda row: geodesic(center, (row['latitude'], row['longitude'])).km,
    axis=1
)
```

```python
df['listing_period'] = (pd.Timestamp.today() - df['created']).dt.days
```

```python
df['created'] = pd.to_datetime(df['created'])
```

```python
df['desc_word_count'] = df['description'].str.split().str.len()
df['price_per_word'] = df['price'] / df['desc_word_count']
```
