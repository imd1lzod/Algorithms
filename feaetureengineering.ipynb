```python
import pandas as pd
import numpy as np

df = pd.read_csv("train.csv")
df
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Order_Year'] = df['Order Date'].dt.year
df['Day_of_the_week'] = df['Order Date'].dt.dayofweek
df['Weekend'] = df['Day_of_the_week'] > 4
```

```python
df2 = pd.read_csv("cities.csv")
df2
```

```python
from geopy.distance import geodesic

center = (41.3, 69.3)
df2['distance_to_center'] = df2.apply(lambda row: geodesic(center, (row['latitude'], row['longitude'])).km, axis=1)
```

```python
df2
```

```python
from matplotlib import pyplot as plt
df2[['latitude','longitude']].plot.scatter(x='latitude', y='longitude', c='blue', s=2, alpha=0.1)
plt.title('Visualization of Latitude and Longitude')
plt.ylabel('longitude')
plt.xlabel('latitude')
# plt.xlim(35, 50)
```
